```python
import pandas as pd

df = pd.read_csv(r'C:\Users\user\Downloads\0306.csv')

df.head()
```

```python
temp = df[['學年度', '設立別','學士班(大學四年制)延修生-男','學士班(大學四年制)延修生-女']]
temp["學士班(大學四年制)延修生-男"] = pd.to_numeric(temp["學士班(大學四年制)延修生-男"])
temp["學士班(大學四年制)延修生-女"] = pd.to_numeric(temp["學士班(大學四年制)延修生-女"])
```

```python
temp['總和']=temp[["學士班(大學四年制)延修生-男","學士班(大學四年制)延修生-女"]].sum(axis=1)
temp['總和']
```

```python
Type = temp.groupby(['學年度'])['總和'].sum().reset_index()
```
