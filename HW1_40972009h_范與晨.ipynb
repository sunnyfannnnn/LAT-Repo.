1. What is the percentage of male and female in this dataset?

```python
gender_male_count = len(data.loc[data['gender'] == 'Male'])
total_count = len(data)
percentage_gender_male = (gender_male_count / total_count) * 100
print('the percentage of male is '+ str(percentage_gender_male) + '%')
gender_female_count = len(data.loc[data['gender'] == 'Female'])
total_count = len(data)
percentage_gender_female = (gender_female_count / total_count) * 100
print('the percentage of female is '+ str(percentage_gender_female) + '%')
```

3. What is the mean and standard deviation of average grades for students studying in academic schools?

```python
stu_grades_mean, stu_grades_std = data.loc[data['type_school'] == 'Academic', 'average_grades'].dropna().mean(), data.loc[data['type_school'] == 'Academic', 'average_grades'].std()

print('The mean is ' + str(stu_grades_mean))
print('The standard deviation is ' + str(stu_grades_std))
```
